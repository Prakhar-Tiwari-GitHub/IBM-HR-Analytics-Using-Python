```python
df.groupby('Gender')['Attrition'].value_counts()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
color_palette = {'Yes':'#1A2421','No':'#79443B'}
color =['#960018', '#9F8170']

sns.countplot(data=df, x='Gender', hue='Attrition', ax=ax1, palette = color_palette)

ax1.bar_label(ax1.containers[0])
ax1.bar_label(ax1.containers[1])
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')
ax1.set_title('Attrition by Gender')

x= df['Gender'].value_counts().index
y= df['Gender'].value_counts().values

ax2.pie(y, labels = x, autopct = "%0.2f%%", pctdistance=0.85, colors=color)
ax2.set_title('Gender Distribution')

plt.suptitle("Attrition by Gender")
plt.tight_layout()

plt.show()
```

```python
gb = df.groupby('Gender')
Attrition_by_Gender = gb['Attrition_No'].sum()
Total_size = gb.size()
Percentage_Attrition_by_Gender = (Attrition_by_Gender/Total_size)*100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
color =['#8F9779', 'Brown'] 

Percentage_Attrition_by_Gender.plot(kind = 'bar', color = '#8F9779', ax = ax1)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
ax1.set_ylabel("Percentage")

Percentage_Attrition_by_Gender.plot(kind = 'pie', autopct = "%0.2f%%", startangle =90, colors=color, ax=ax2)
ax2.set_title(" Percent Distribution of [% of Attrition in Gender]")

plt.suptitle(" % Attrition by Gender", fontsize=16)
plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 9))
custom_palette = {'Yes': 'grey', 'No': '#A52A2A'}
colors = ['#708238', '#27AE60', '#79443B', '#3498DB', '#8F9779', '#960018']

sns.countplot(data=df, x='EducationField', hue='Attrition', ax=ax1, palette=custom_palette)

ax1.bar_label(ax1.containers[0])
ax1.bar_label(ax1.containers[1])
ax1.set_xlabel('Education Fields')
ax1.set_ylabel('Count')

x= df['EducationField'].value_counts().index
y= df['EducationField'].value_counts().values

ax2.pie(y, labels = x, autopct = "%0.2f%%", pctdistance=0.85, colors=colors)

plt.tight_layout()
fig.suptitle('Attrition by Education Fields', fontsize=16)
plt.subplots_adjust(top=0.93,)


plt.show()
```

```python
gb1 = df.groupby('EducationField')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))
colors = ['#EFDECD', '#3498DB', '#79443B', '#27AE60', '#FFFDD0', '#960018']

attrition_counts = gb1['Attrition_No'].sum()
total_counts = gb1.size()
attrition_percentage = (attrition_counts / total_counts)*100

attrition_percentage.plot(kind='bar', color = '#3D2B1F', ax=ax1)
ax1.set_ylabel("Percentage")
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())

attrition_percentage.plot(kind='pie', autopct='%1.2f%%', startangle=89,
                          ax=ax2, colors=colors)
ax2.set_title(" Percent Distribution of [% of Attrition in Eduaction Fields]")

fig.suptitle('Attrition % in Education Fields', fontsize=16)
plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
color_palette = {'Yes': '#9F8170', 'No':'#D21F3C'}
color =['#8F9779', 'Brown','#79443B' ] 

x = df['BusinessTravel'].value_counts().values
y = df['BusinessTravel'].value_counts().index

ax = sns.countplot(data= df, x= 'BusinessTravel', hue= 'Attrition', ax=ax1, palette= color_palette)
plt.pie(x, labels= y, autopct='%0.2f%%', startangle= 90, colors=color )
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.tight_layout()
ax1.set_title("Attrition in Business Travel")
ax2.set_title("% of different Business Travel")

plt.show()
```

```python
gb = df.groupby('BusinessTravel')
Total_count = gb.size()
Attrition_in_Business_Travel = gb['Attrition_No'].sum()
Attrition_Per_in_Business_Travel = (Attrition_in_Business_Travel/Total_count)*100
color =['#5F9EA0', '#8FBC8F', '#FFE4C4'] 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (13,6))

Attrition_Per_in_Business_Travel.plot( kind = 'bar', color = '#2F4F4F', ax = ax1)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
ax1.set_ylabel("Percentage")

Attrition_Per_in_Business_Travel.plot(kind = 'pie', autopct = '%0.2f%%', startangle = 90, colors = color)
ax2.set_title(" Percent Distribution of [% of Attrition in Business Travel]")

plt.tight_layout()
fig.suptitle("% Attrition In Each Business Travel", fontsize = 16)
plt.subplots_adjust(top =0.88)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 4))
color_palette = {'Yes':'#DB7093','No':'#6B8E23'}
colors = ['#DDA0DD','#FAFAD2','#F08080']
x = df['Department'].value_counts().values
y = df['Department'].value_counts().index

ax = sns.countplot(data= df, x= 'Department', hue= 'Attrition', ax=ax1, palette= color_palette)
plt.pie(x, labels= y, autopct='%0.2f%%', startangle= 90, colors = colors )
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.tight_layout()
ax1.set_title("Attrition in Department")
ax2.set_title("% of Each Department")

plt.show()
```

```python
gb = df.groupby('Department')
Total_counts = gb.size()
Attrition_by_Department = gb['Attrition_No'].sum()
Attrition_Per_by_Department = (Attrition_by_Department/Total_counts)*100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
colors = ['#6495ED','#483D8B','#BDB76B']

ax1 = Attrition_Per_by_Department.plot(kind = 'bar', color ='#8B0000', ax = ax1 )
ax1.bar_label(ax1.containers[0])
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
ax1.set_ylabel("Percentage")

Attrition_Per_by_Department.plot(kind='pie', autopct ='%0.2f%%', colors = colors)
ax2.set_title(" Percent Distribution of [Attrition % in Each Department]")  

plt.tight_layout()
plt.suptitle("Attrition % in Each Department", fontsize = 16)
plt.subplots_adjust(top=0.91)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))
color_palette = {'Yes':'#483D8B', 'No':'#5F9EA0'}
color =['#7FFFD4', '#8FBC8F', '#FFE4C4','#BDB76B' ]
sns.histplot(data = df , x = 'RelationshipSatisfaction', hue = 'Attrition', palette = color_palette, ax = ax1)
for container in ax1.containers:
    ax1.bar_label(container)
    
x = df['RelationshipSatisfaction'].value_counts().values
y = df['RelationshipSatisfaction'].value_counts().index
plt.pie( x, labels = y,autopct = '%0.2f%%', startangle = 90, colors = color)

plt.tight_layout()
ax1.set_title("Attrition By Relationship Satisfaction")
ax2.set_title("% Distribution By Realtionship Satisfaction")
plt.show()
```
